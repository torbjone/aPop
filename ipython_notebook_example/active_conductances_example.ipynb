# Single synaptic input
To run a NEURON simulation with a cortical pyramidal cell that receives a
single synaptic input, simply run the code block bellow. The Local Field Potential is calculated and
plotted.

*Parameters*:
---------------
* **synaptic_z_pos**: (float, int) - position along the apical dendrite where the synapse is inserted.
* **conductance_type**: (str) - Either 'active', 'passive', 'Ih' or 'Ih_frozen'. If 'active' all original ion-channels are included, if 'passive' they are all removed. 'Ih' will include a single active Ih channel. 'Ih_frozen' results in
    a passive cell with the passive membrane conductance scaled to include
    the increased membrane conductance that comes with the Ih conductance
* **weight**: (float) - Strength of synaptic input.
* **holding_potential**: (int, float, None) - If a number is given (e.g. - 70), the passive leak reversal potential
    is modified to make the given number the resting potential of the cell,
    i.e. the resting potential will be uniform.
    Defaults to None, which does not enforce a uniform potential.

```python
example_synapse(synaptic_z_pos=1000, 
                conductance_type='active', 
                weight=-0.0005, 
                holding_potential=-70)
```
